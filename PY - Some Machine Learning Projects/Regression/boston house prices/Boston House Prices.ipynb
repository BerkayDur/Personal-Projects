```python
from sklearn.datasets import load_boston

data = load_boston()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
X = data["data"]
y = data["target"]
X
```

```python
pd = pd.DataFrame(np.c_[X,y],columns=np.r_[data["feature_names"], ["target"]])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, pd_train, pd_test = train_test_split(X, y, pd, test_size=0.2)
```

```python
train_corr = pd_train.corr()
```

```python
from pandas.plotting import scatter_matrix

features = ["target","RM","LSTAT","PTRATIO","INDUS","TAX","NOX"]

scatter_matrix(pd_train[features], figsize=(20,16))
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
X_train_scaled = std_scaler.fit_transform(X_train)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline


sgd_1 = SGDRegressor(random_state=42)
sgd_2 = Pipeline([
    ("poly_2", PolynomialFeatures(degree=2)),
    ("sgd", SGDRegressor(random_state=42)),
])
sgd_4 = Pipeline([
    ("poly_4", PolynomialFeatures(degree=4)),
    ("sgd", SGDRegressor(random_state=42)),
])
sgd_8 = Pipeline([
    ("poly_8", PolynomialFeatures(degree=8)),
    ("sgd", SGDRegressor(random_state=42)),
])
```

```python
from sklearn.model_selection import learning_curve

train_size, train_scores, test_scores = learning_curve(sgd_1, X_train, y_train, cv=2, train_sizes=np.linspace(0.001,1.0,1000),
                                                      scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
```

```python
train_size, train_scores, test_scores = learning_curve(sgd_2, X_train, y_train, cv=2, train_sizes=np.linspace(0.001,1.0,1000),
                                                      scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
```

```python
train_size, train_scores, test_scores = learning_curve(sgd_4, X_train, y_train, cv=2, train_sizes=np.linspace(0.001,1.0,1000),
                                                      scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
```

```python
train_size, train_scores, test_scores = learning_curve(SGDRegressor(penalty="elasticnet", alpha=30000, l1_ratio=0.5),
                                                       X_train, y_train, cv=2, train_sizes=np.linspace(0.01,1.0,100),
                                                       scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
plt.axis([0,220,0,15])
```

```python
from sklearn.ensemble import RandomForestRegressor

train_size, train_scores, test_scores = learning_curve(RandomForestRegressor(random_state=42),
                                                       X_train, y_train, cv=2, train_sizes=np.linspace(0.01,1.0,100),
                                                       scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
plt.axis([0,220,0,15])
```

```python
train_size, train_scores, test_scores = learning_curve(RandomForestRegressor(random_state=42, n_estimators=50),
                                                       X_train, y_train, cv=2, train_sizes=np.linspace(0.01,1.0,100),
                                                       scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
plt.axis([0,220,0,15])
```

```python
train_size, train_scores, test_scores = learning_curve(RandomForestRegressor(random_state=42, n_estimators=20),
                                                       X_train, y_train, cv=2, train_sizes=np.linspace(0.01,1.0,100),
                                                       scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
plt.axis([0,220,0,15])
```

```python
train_size, train_scores, test_scores = learning_curve(RandomForestRegressor(random_state=42, n_estimators=10),
                                                       X_train, y_train, cv=2, train_sizes=np.linspace(0.0001,1.0,1000),
                                                       scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
plt.axis([0,220,0,15])
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{
        "n_estimators": [8, 16, 32, 64, 128],
        "bootstrap": [True, False],
        "max_samples": [None, 8, 16, 32, 64],
        "random_state": [42]
    }]


rscv = GridSearchCV(RandomForestRegressor(), param_grid=param_grid, scoring="neg_root_mean_squared_error",
                    cv=3, return_train_score=True)
```

```python
rscv.fit(X_train, y_train)
```

```python
rscv.cv_results_
```

```python
for i,j in list(zip(-rscv.cv_results_["mean_test_score"],rscv.cv_results_["params"])):
    print(i,j)
```

```python
model = rscv.best_estimator_
```

```python
train_size, train_scores, test_scores = learning_curve(model,
                                                       X_train, y_train, cv=2, train_sizes=np.linspace(0.01,1.0,100),
                                                       scoring="neg_root_mean_squared_error", random_state=42)

plt.plot(train_size,-train_scores[:,0], "r-+", label="Train")
plt.plot(train_size,-test_scores[:,0], "b-", label="Validation")
plt.legend()
plt.axis([0,220,0,15])
```

```python
Final = Pipeline([
    ("scaler", StandardScaler()),
    ("model",model),
])
```

```python
Final.fit(X_train, y_train)
```

```python
X_test_predict = Final.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

final_error = mean_squared_error(y_test, X_test_predict, squared=False)
```

```python
final_error
```

```python
rel_error = 100 * abs(y_test - X_test_predict) / y_test 
```

```python
print("Average Percentage Error: +-", rel_error.mean(),"%")
print("Standard Deviation: ", rel_error.std(), "%")
```
